Reading *guns.csv* file and creating a list of lists with each row in the file a sub-list in the outer list

```python
import csv
data = list(csv.reader(open("guns.csv", "r")))
print (data[0:10])
```

Separating the header row from the records 

```python
header = data[0]
data = data[1:]
print (header)
```

Retrieving the years for which the data is provided:

```python
years = {row[1] for row in data}
print (years)
```

The data has been provided for the years **2012**, **2013**, and **2014**. 

Retrieving the intents behind the gun deaths listed:

```python
intents = {row[3] for row in data}
print (intents)
```

The intents behind the gun deaths recorded are **Undetermined**, **Homicide**, **Suicide**, **Accidental**, and **NA**.  

Retrieving the races included in the data:

```python
races = {row[7] for row in data}
print (races)
```

The races recorded in the data: **Native American/Native Alaskan**, **Black**, **White**, **Asian/Pacific Islander**, and **Hispanic**. 

Retrieving the various places where the gun death incidents were reported:

```python
places = {row[9] for row in data}
print (places)
```

The gun death incidents have occurred **at Farms, on Industrial/Construction sites, in Trade and service areas, on streets, at schools and educational institutions, at homes, and at Residential institutions.** Some locations are **unspecified** and others mentioned as **NA**. **Sports** related gun deaths have also been recorded.   

Retrieving the educational status of the victims:

```python
educational_status = {row[10] for row in data}
print (educational_status)
```

The **educational status** of the victim coded as follows:   
1 - Less than High School    
2 - Graduated from High School or equivalent  
3 - Some College  
4 - At least graduated from college  
5 - Not Available   
NA - Not Applicable   

Finding the number of gun deaths in each of the given years:

```python
year_counts = {}
for row in data:
    year = row[1]
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
print (year_counts)
```

Gun deaths recorded in the years **2012**, **2013**, and **2014** are **33563**, **33636**, and **33599**, respectively. We can infer that there hasn't been any significant change in the number of gun deaths from the year 2012 to 2014. 


Now we will find the number of male and female victims of gun deaths in each of the given three years: 

```python
sexes = {row[5] for row in data}
print (sexes)
```

```python
def year_sex_count(year):
    count = {'M':0, 'F':0}
    for row in data:
        if row[1] == year:
            gender = row[5]
            count[gender] += 1
    return count
```

```python
sex_count_2012 = year_sex_count('2012')
print (sex_count_2012)

sex_count_2013 = year_sex_count('2013')
print (sex_count_2013)

sex_count_2014 = year_sex_count('2014')
print (sex_count_2014)
```

We can roughly say from the above numbers that the number of male casualties of gun deaths is way more than the number of female casualties. However, to come to such a conclusion, we will still need data on the female and the male population off the total population of the US.  
From the number of female casualties, we see that the number has slightly risen in 2013 to 4842 against the 4725 recorded the previous year.  

```python
sex_counts = {'M': sex_count_2012['M'] + sex_count_2013['M'] + sex_count_2014['M'], 'F': sex_count_2012['F'] + sex_count_2013['F'] + sex_count_2014['F']}
print sex_counts
```

The total number of male and female casualties of gun deaths between 2012 and 2014 are **86349** and **14449**, respectively. 

Finding the oldest victims of gun deaths:

```python
#replacing 'NA' and '0' values for the age column with None
for row in data:
    if row[6] == 'NA' or row[6] == '0':
        row[6] = None

#Fetching the age of all the listed victims and storing it in a set
ages = {int(row[6]) if row[6] is not None else None for row in data}

#finding the age of the oldest victim
max_age = 0
for age in ages: 
    if age is not None and age > max_age:
        max_age = age
        
#fetching the details of the oldest victims of gun deaths     
oldest_victims = filter(lambda row: row[6] is not None and int(row[6])==max_age, data)
print (oldest_victims)
```

The oldest victim of gun deaths is reported to be a **Black Male**, aged **107**. The intent behind the gun violence is **Homicide**. The victim was killed in his home in September 2013.  

Finding the youngest victims of gun deaths: 

```python
#finding the age of the youngest victims
min_age = iter(ages).next()
for age in ages:
    if age is not None and age < min_age:
        min_age = age

#fetching the details of the youngest victims of gun deaths     
youngest_victims = filter(lambda row: row[6] is not None and int(row[6])==min_age, data)
print (youngest_victims[0:5])
```

```python
#finding the intent behind the gun deaths of the youngest victims
intent_youngest_victims = {row[3] for row in youngest_victims}
print intent_youngest_victims
```

```python
#the number of youngest victims of gun deaths
num_youngest_victims = len(youngest_victims)
print num_youngest_victims
```

```python
#number of homicide and accidental cases off the total number of cases involving the youngest victims
count_intent_youngest_victims = {intent:0 for intent in intent_youngest_victims}

for row in youngest_victims:
    intent = row[3]
    count_intent_youngest_victims[intent] += 1

print count_intent_youngest_victims    
        
```

```python
#finding the number of homicide cases, off the 34, by race of the youngest casualities
race_homicide_count_youngest_victims = {}
for row in youngest_victims:
    intent = row[3]
    race = row[7]
    if intent == 'Homicide':
        if race not in race_homicide_count_youngest_victims:
            race_homicide_count_youngest_victims[race] = 1
        else:
            race_homicide_count_youngest_victims[race] += 1
print race_homicide_count_youngest_victims
```

A total of **38** cases have been recorded where the victims are **aged 1**; the youngest casualties of gun deaths. Off the 38 cases, **34** are **homicides** and **4** are **accidental deaths**. And off the 34 homicide victims, **16** were of the *Black* descent, 9 White, 7 Hispanic, and 2 Asian/Pacific Islander.

Now we will find the number of gun deaths by place of the incident:

```python
place_count = {}
for row in data:
    place = row[9]
    if place not in place_count:
        place_count[place] = 1
    else:
        place_count[place] += 1
print place_count
```

The numbers suggest that by far most incidents of gun deaths were reported to have occurred at home. 

```python
#funtion to count by place and intent
def place_intent_count(place):
    count = {}
    for row in data: 
        if row[9] == place:
            intent = row[3]
            if intent not in count:
                count[intent] = 1
            else:
                count[intent] += 1
    return count
```

```python
#the count by intent of the gun deaths at home
home_intent_count = place_intent_count('Home')
print home_intent_count
```

And most deaths that occurred at **home** were a result of the victims taking their own lives. The number of suicides recorded is **45415**, followed by Homicide at **13613**.   

```python
#the count by intent of gun deaths at schools and educational institutions
school_intent_count = place_intent_count('School/instiution')
print school_intent_count
```

The gun deaths that occurred at **schools** and **educational institutions** were largely due to **Homicide (361 cases)** and **Suicide (295 cases)**. 

```python
#funtion to count by the race of the victim, the intent, and the place
def place_intent_race_count(place, intent):
    count={}
    for row in data:
        if row[9] == place and row[3] == intent:
            race = row[7]
            if race not in count:
                count[race] = 1
            else:
                count[race] += 1
    return count
```

```python
home_suicide_race_count = place_intent_race_count('Home', 'Suicide')
print home_suicide_race_count
```

Off the **45415** cases of **suicides** reported at **Home**, **39993**, i.e. **88%**, of the victims were **Whites**.  

```python
home_homicide_race_count = place_intent_race_count('Home', 'Homicide')
print home_homicide_race_count
```

Off the **13613** cases of **Homicides** at **Home**, **6413** of the victims were **Blacks**, **4905 Whites**, and **1949 Hispanics**. 

```python
school_homicide_race_count = place_intent_race_count('School/instiution', 'Homicide')
print school_homicide_race_count
```

Off the **361** cases of **homicides**, the foremost cause of gun deaths, at **schools** and **educational institutions**, **201** of the victims were Blacks. 

```python
school_suicide_race_count = place_intent_race_count('School/instiution', 'Suicide')
print school_suicide_race_count
```

Off the **295** cases of **Suicides** reported at **schools** and **educational institutions**, **239**, i.e. **81%**, of the victims were **Whites**.

We will read another file *census.csv* that has data pertaining to the population of different races off the total population of the US. From that data, we will compute the number of gun deaths per 100000 people in each race.  

```python
census = list(csv.reader(open("census.csv")))
print census
print '\n'

#separating the header row and assigning it to a list
census_header = census[0]
print census_header
print '\n'

census = census[1]
print census
```

We will manually create a mapping of the race and the population count of that race from the census. Each key in the race_counts will be the key in the dictionary, as the race names in the census and the gun deaths data differ.  

```python
mapping = {'Asian/Pacific Islander': int(census[15])+int(census[14]), 'White':int(census[10]), 'Hispanic':int(census[11]), 'Black':int(census[12]), 'Native American/Native Alaskan': int(census[13])} 
print mapping
```

```python
race_count_per_hundredk = {race: float(count*100000)/mapping[race] for race, count in race_counts.items()}
print (race_count_per_hundredk)
```

The above numbers depict the *rate of gun deaths per 100000 people of each racial category*. 

```python
#homicide counts by race
homicide_race_counts = {}
for row in data:
    race = row[7]
    intent = row[3]
    if intent == 'Homicide':
        if race in homicide_race_counts:
            homicide_race_counts[race] += 1
        else:
            homicide_race_counts[race] = 1
print homicide_race_counts
```

```python
homicide_race_per_hundredk = {race: float(count*100000)/mapping[race] for race,count in homicide_race_counts.items()}
print homicide_race_per_hundredk
```
